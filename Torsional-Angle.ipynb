```python
import math
# PROBLEM: Given 4 points (A,B,C,D), return the angle between planes (A,B,C) and (B,C,D) in degrees (phi)
# https://www.hackerrank.com/challenges/class-2-find-the-torsional-angle/problem?isFullScreen=true

# APPROACH:
# Define the attributes (x,y,z) for the Points Class
# Step 1: Define method to subtract two points
# Step 2: Define method to calculate dot product
# Step 3: Define method to calculate cross product

class Points(object):
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, no):
        x = self.x - no.x
        y = self.y - no.y
        z = self.z - no.z
        return Points(x,y,z)

    def dot(self, no):
        # ==== Dot Product Example ====== #
        # A.B = A[0]*B[0] + A[1]*B[1] + A[2]*B[1] ...
        x = self.x * no.x
        y = self.y * no.y
        z = self.z * no.z
        # return Points(x, y, z)
        return x + y + z
        

    def cross(self, no):
        # ==== CROSS PRODUCT ==== #
        # EXAMPLE: A x B 
        # x = (Ay * Bz) - (Az * By)
        # y = (Az * Bx) - (Ax * Bz)
        # z = (Ax * By) - (Ay * Bx)
        x = self.y * no.z - self.z * no.y
        y = self.z * no.x - self.x * no.z
        z = self.x * no.y - self.y * no.x
        return Points(x,y,z)
            
    def absolute(self):
        return pow((self.x ** 2 + self.y ** 2 + self.z ** 2), 0.5)

if __name__ == '__main__':
    points = list()
    for i in range(4):
        a = list(map(float, input().split()))
        points.append(a)

    a, b, c, d = Points(*points[0]), Points(*points[1]), Points(*points[2]), Points(*points[3])
    x = (b - a).cross(c - b)
    y = (c - b).cross(d - c)
    angle = math.acos(x.dot(y) / (x.absolute() * y.absolute()))

    print("%.2f" % math.degrees(angle))
```
